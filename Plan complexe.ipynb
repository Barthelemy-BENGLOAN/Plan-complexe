# Le plan complexe 
*Ce notebook est fait pour comprendre les grandes lignes de la necessité du plan complexe illuster avec une petite application*

 **<u>Repère 2D et ses limites</u>**

Jusqu’à la Renaissance, toute fonction et en particulier les polynômes du second degr se représentait sur un plan cartésien à deux axes (abscisse : \(x\), ordonnée : \(f(x)\)). <br><br>Lorsque la parabole définie par  
$$
f(x) = a x^2 + b x + c
$$  
ne coupe pas l’axe horizontal, on en concluait qu’il n’existait aucune solution réelle à l’équation  
$$
a x^2 + b x + c = 0.
$$  

En effet $\mathbb{R}$  n’est pas clos pour l’opération de racine carrée : tirer la racine d’un nombre négatif fait inévitablement sortir du cadre des réels. <br><br> Cette impossibilité géométrique conduisait les mathématiciens à qualifier certains cas d’irréductibles.
<br><br><br>

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 + 1

x = np.linspace(-10, 10, 400)
y = f(x)

fig, ax = plt.subplots()

ax.plot(x, y, linewidth=2)
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
ax.set_title("f(x) = x² + 1")

ax.spines["left"].set_position("zero")
ax.spines["bottom"].set_position("zero")
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
plt.tight_layout()
plt.show()
```

*On remarque que dans cet exemple la parabole ne coupe pas l'axe horizontale*

**<u>Naissance du plan complexe</u>**

C'est au XVIIᵉ siècle que Cardan s’est confronté à l’apparition de racines carrées de nombres négatifs en résolvant des équations du troisième degré. Imprégné de l’idée que tout objet mathématique doit se représenter géométriquement, il met ces quantités entre parenthèses. Un siècle plus tard, Descartes baptise ces inconnues imaginaires sans leur accorder de réalité propre.

En 1796, Caspar Wessel propose une interprétation géométrique, il se dit que si multiplier par \(-1\) correspond à une rotation de 180 degrés autour de l’origine de notre notre plan complexe. <br><br> Alors poser
 
$$
\sqrt{-1} = i
$$  
revient donc à imaginer une rotation de 90 degrés, c’est‑à‑dire un point à mi‑chemin entre l’axe réel et un nouvel axe perpendiculaire. Gauss nomme cet espace le plan complexe, un espace vectoriel réel de dimension 2 où l’axe imaginaire complète l’axe réel.

**<u>Retour sur notre exemple</u>**

Ici pour un polynome d'ordre 2 on a donc que lorsque le discriminant est négatif (comme c'est le cas dans notre exemple) nous n'avons pas de solution dans $\mathbb{R}$ <br><br> En revanche dans $\mathbb{C}$

$$
x = \frac{-b \pm \sqrt{\Delta}}{2a}
    = \frac{-b}{2a} \pm i\,\frac{\sqrt{|\Delta|}}{2a}.
$$    
Ainsi sur  $\mathbb{R}$, aucune solution mais maintenant sur  $\mathbb{C}$ on obtient  

$$
x = \pm i,
$$  


```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-2, 2, 200)
y = np.linspace(-2, 2, 200)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y 

F = np.abs(f(Z))

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, F, cmap='viridis')

ax.set_xlabel('Partie réelle')
ax.set_ylabel('Partie imaginaire')
ax.set_zlabel('|f(z)|')
ax.view_init(elev=10, azim=310)

plt.show()
```

*On visualise donc maintenant que notre fonction admet bien 2 solutions dans le plan complexe*
